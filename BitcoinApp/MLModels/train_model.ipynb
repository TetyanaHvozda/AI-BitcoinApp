```python
# train_model.ipynb
import pandas as pd
from sklearn.linear_model import LinearRegression
import coremltools as ct
```

```python
# Load the saved CSV file
df = pd.read_csv("btc_prices.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
df.set_index("timestamp", inplace=True)
```

```python
# Create a shifted column for previous day's price
df["previous_price"] = df["price"].shift(1)
df.dropna(inplace=True)
```

```python
# Prepare features and labels
X = df["previous_price"].values.reshape(-1, 1)
y = df["price"].values
```

```python
# Train a simple linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Convert the model to CoreML
mlmodel = ct.converters.sklearn.convert(
    model,
    input_features=["previous_price"],         
    output_feature_names="predicted_price"       
)
```

```python
# Save to .mlmodel file
mlmodel.save("BitcoinPricePredictor.mlmodel")
print("✅ Model saved as BitcoinPricePredictor.mlmodel")
```
